```python
import pint, numpy as np

# Create a UnitRegistry instance. This will load the default units.
u = pint.UnitRegistry()

c = u.m / u.s
k = u.J / u.K
ln_Lambda = 10 * u.dimensionless
Z = 1 * u.dimensionless
e = 1 * u.coulomb
T = 1 * u.K
me = 1 * u.gram

# Desired units are m^2/s
eta = np.sqrt(2) * c**2 * ln_Lambda * Z * e**2 * np.sqrt(me)/ (12 * (np.pi * k * T) ** 1.5)
print(eta.to_base_units().to_compact())
```
